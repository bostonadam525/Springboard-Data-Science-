```python
# what need to do is sort these 32 boroughs from the list we made
# lets upload a text file of these boroughs to work with
path = '/content/drive/MyDrive/Springboard_Data_Science/London Housing Case Study/actual_32_boroughs.txt'


# define a function to clean and process this text file

def process_text(path):
  global actual_boroughs # make this a global variable outside of the function
# open the text file
  with open(path, 'r') as file:
    actual_boroughs = file.read()
    # remove new lines
    actual_boroughs = actual_boroughs.split('\n')
    # sort in order
    actual_boroughs = sorted(actual_boroughs)
    # replace 'and' with '&' so we can match the strings in the dataframe
    actual_boroughs = [sub.replace(' and ',' & ') for sub in actual_boroughs]

    return actual_boroughs

  # close the file
  file.close()

```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great!

***Hint***: This section should test the skills you acquired in:

```python
### Let's write a function to calculate the price ratio
def create_price_ratio(d):
  y_1998 = float(d['Average_price'][d['Year'] == 1998])
  y_2018 = float(d['Average_price'][d['Year'] == 2018])
  price_ratio = [np.mean(y_1998)/np.mean(y_2018)]
  return price_ratio
```

```python
# lets test the function
create_price_ratio(df_no_london_mean[df_no_london_mean['London_Borough'] == 'Barking & Dagenham'])
```

### Write a function to do this

```python
## We want to make a dictionary to store our data for the price_ratio for our data and then turn it into a dataframe
def price_ratio_loop(x):
  '''This function will take in a dataframe, then iterate through the unique elements of the London_Borough column and apply the create_price_ratio function and append
  it to a dictionary '''
  final_dict = {}

  for b in x['London_Borough'].unique():
      boroughs = x[x['London_Borough'] == b]
      final_dict[b] = create_price_ratio(boroughs)

  return final_dict
```

```python
## apply the price_ratio_loop function to each dataframe
## let's apply it to the df_no_london_mean first and set it to a variable we can use
df_no_london_dict = price_ratio_loop(df_no_london_mean)

# print dict
print(df_no_london_dict)
```

```python
## now apply the function to the df_with_london_mean df
df_with_london_dict = price_ratio_loop(df_with_london_mean)

# print dict
print(df_with_london_dict)
```
