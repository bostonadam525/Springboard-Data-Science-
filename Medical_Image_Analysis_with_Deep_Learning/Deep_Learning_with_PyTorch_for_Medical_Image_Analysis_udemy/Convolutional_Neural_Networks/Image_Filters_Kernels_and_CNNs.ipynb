## Filters

### What are filters?
* Filters are an **image kernel**
    * defined: a small matrix applied to an entire image.
* Certain filter types are well known such as a **blur filter**
    * The "blur filter" or "image kernel" is actually a 3x3 matrix of specific values that when applied will result in a blurry image (source: Pieran Data)

## Stride Distance
* The movement of the filter across the image/kernels or matrix.
* Example: setosa.io/ev/image-kernels/
* Image values range from 0 to 255.
    * Darker images => closer to 0
    * Lighter images => closer to 255

* Kernel Application "striding" across an image is called "Convolutions"
    * A neural network will do this using weights and biases randomization.
    * A pre-set kernel/filter such as the "blur" kernel are pre-set weights.
    * A neural network will essentially sample different weights to find the optimal filters/kernels in a machine learning task (i.e. classification)

# Convolutional Kernels
* "filters" or "kernels" are commonly referred to as **convolution kernels**
* Process of passing the filters/kernels over an image is a **convolution**

## How do we handle missing filters/kernels?
* This most commonly occurs when the sliding window or filter hits the edge of the image where there are no values.
* Most common way to handl this is with "padding" similar to how we handle NLP text sequences.
* We would pad with 0 floats < 1.

# Convolutional Neural Network Architecture

## Convolutional Layers
* If we were to always us ANN's (artificial neural nets) for image analysis or deep learning:
1. It would require a LARGE AMOUNT of parameters (e.g. >100,000 for 28x28 images)
2. All 2D information would be lost by flattening out the image just to fit the vector into the NN.
3. ANN's only perform well where the image data are all similar (e.g. MNIST datast where the numbers or pixels are all perfectly centered)
    * An ANN would not be able to generalize on unseen images where the numbers are NOT centered.

## What is a convolutional layer?
* Convoluational layers are created when we apply **multiple image filters** to the input image data.
* We then train the neural network to determine the optimal weights for the convolutional filters.

## What is local convolutional local connectivity?
* CNNs reduce model params by using **local connectivity**.
* Not all neurons in the network will be fully connnected.
* Thus, neurons are only connected to a subset of local neurons in the next layer => these become the filters.

### Breakdown of local connectivity in CNNs
* Fully connected NN's usually have multiple connections between neurons.
* This is the major issue with an ANN, too many params=too much time to train=too complex.

# Pooling Layers
* Even though we have **local connectivity** which is supposed to help reduce the number of filters and parameters needed, when working with color images you can still have 10s or 100s of filters with more parameters.
* Enter => **POOLING LAYERS**
    * Pooling layers accept convolutional layers as input.
    * Convolutional layers will have many, many filters/tensors.
    * Pooling layers can help reduce this.

## Pooling Layer details
* Also called "subsampling".
1. Max Pooling Layer
  * Window 2x2, Stride 2
  * Slide window and grab "max" value of each window.
  * Some information loss occurs.
  * In the red filter below we can see the max is "3" which is retained in the final matrix or tensor.
image.png

2. Average Pooling Layer
* Same process as above, but average out values in each window.

3. The pooling layer will remove a lot of information from the convolutional layers.
    * As an example, a small pooling layer with a "kernel" or "filter" of 2x2 with stride of 2 will remove 75% of the input convolutional data.

# Dropout
* This is another common technique deployed to avoid overfitting the CNN neural network. This is commonly used in hyperparameter tuning for any type of deep learning algorithm.
* Concept Review:
    1. Droput is a regularization technique to prevent overfitting.
    2. During network training, units will be "randomly dropped" or "turned off" along with the neuron connection.
    3. This helps units from "co-adapting" too much.
